```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data=pd.read_csv("Superstore.csv")
```

```python
data.head(10)
```

```python
data.describe()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data.fillna(0 , inplace=False)
```

```python
data["Order Date"] =pd.to_datetime(data["Order Date"]).dt.month_name()
data["Order Date"]
```

```python
data["Profit"]
```

```python
data_profit=data.groupby("Order Date")["Profit"].sum().reset_index()
data_profit
```

```python
data["Discount"]
```

```python
data_discount=data.groupby("Order Date")["Discount"].sum().reset_index()
data_discount
```

```python
data_category=data.groupby("Discount")["Profit"].mean().reset_index().sort_values("Discount",ascending=True)
data_category
```

```python
plt.title("Profit")
sns.barplot(data_profit, x='Order Date', y='Profit')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.title("Discount")
sns.barplot(data_discount, x='Order Date', y='Discount')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.title("Discount&Profit")
sns.lineplot(data_category, x='Discount', y='Profit')
plt.xticks(rotation=45)
plt.show()
```

```python
segment = data.groupby("Segment")[["Sales","Profit"]].sum()
```

```python
segment
```

```python
plt.figure(figsize=(12,6))
segment.plot(kind="bar")
plt.title("Segment")
plt.xlabel("Segment")
plt.ylabel("Sales , Profit")
plt.show()
```

```python
sns.pairplot(data, vars=['Sales','Quantity','Discount','Profit'], hue='Sales')
plt.show()
             
```

```python
numbers= data[['Sales','Quantity','Discount','Profit','Postal Code','Row ID']]
```

```python
corr_matrix=numbers.corr()
corr_matrix
```

```python
plt.figure(figsize=(8,8))
sns.heatmap(corr_matrix,annot =True,cmap='coolwarm',fmt=".2f")
```

```python
for col in numbers.columns:
    sns.boxplot(x=numbers[col],showmeans=True)
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data= pd.read_csv('Superstore.csv', encoding='latin-1')
data.head(10)
```

```python
data.tail(6)
```

```python
data.sample(5)
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data.fillna(0 ,inplace=False)
```

```python
data.describe()
```

```python
data["Order Date"]=pd.to_datetime(data["Order Date"])
data["Order Date"]
```

```python
data["Ship Date"]=pd.to_datetime(data["Ship Date"])
data["Ship Date"]
```

```python
data["order month"] =pd.to_datetime(data["Ship Date"]).dt.month_name()
data["order month"]
```

```python
data["year"]=pd.to_datetime(data["Order Date"]).dt.year
data["year"]
```

```python
mothly_salse=data.groupby("order month")['Sales'].sum().reset_index().sort_values('Sales',ascending=False)
mothly_salse
```

```python

plt.title("Monthly Sales")
sns.lineplot(mothly_salse, x='order month', y='Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
category_salse=data.groupby("Category")['Sales'].sum()
category_salse
```

```python
plt.title("sales by each Category")
colors=sns.color_palette("pastel")[0:len(category_salse)]
plt.pie(category_salse, labels=category_salse.index, colors=colors, autopct='%1.1f%%')
plt.show()
```
